```python
#imports
import json
import pandas as pd
import datetime as dt
import numpy as np
```

## "V" file

```python
#open file
vfile = open('entire_state_v.lst')
vfile_lines = vfile.readlines()
```

```python
#test
sample_vfile_line = (vfile_lines[0])
print(sample_vfile_line)
```

```python
sample_vfile_line = (vfile_lines[1])
print(sample_vfile_line)
```

```python
sample_vfile_line = (vfile_lines[2])
print(sample_vfile_line)
```

```python
#separate each line into a dictionary
vfile_rows = []
for line in vfile_lines[:5]:
    row = {}
    row['last'] = line[0:35].strip()
    row['first'] = line[35:55].strip()
    row['middle'] = line[55:75].strip()
    row['suffix'] = line[75:78].strip()
    row['birth_year'] = line[78:82].strip()
    row['gender'] = line[82:83].strip()
    row['reg_date'] = line[83:91].strip()
    row['house_number_character'] = line[91:92].strip()
    row['residence_street_number'] = line[92:99].strip()
    row['house_suffix'] = line[99:103].strip()
    row['predirection'] = line[103:105].strip()
    row['street_name'] = line[105:135].strip()
    row['street_type'] = line[135:141].strip()
    row['suffix_direction'] = line[141:143].strip()
    row['residence_extension'] = line[143:156].strip()
    row['city'] = line[156:191].strip()
    row['state'] = line[191:193].strip()
    row['zip'] = line[193:198].strip()
    row['mail_add_1'] = line[198:248].strip()
    row['mail_add_2'] = line[248:298].strip()
    row['mail_add_3'] = line[298:348].strip()
    row['mail_add_4'] = line[348:398].strip()
    row['mail_add_5'] = line[398:448].strip()
    row['voter_id'] = line[448:461].strip()
    row['county_code'] = line[461:463].strip()
    row['jurisdiction'] = line[463:468].strip()
    row['ward_precinct'] = line[468:474].strip()
    row['school_code'] = line[474:479].strip()
    row['state_house'] = line[479:484].strip()
    row['state_senate'] = line[484:489].strip()
    row['us_congress'] = line[489:494].strip()
    row['county_commissioner'] = line[494:499].strip()
    row['village_code'] = line[499:504].strip()
    row['village_precinct'] = line[504:510].strip()
    row['school_precinct'] = line[510:516].strip()
    row['permanent_absentee_ind'] = line[516:517].strip()
    row['status_type'] = line[517:519].strip()
    row['uocava_status'] = line[519:520].strip()
    
    vfile_rows.append(row)
```

```python
vfiledf = pd.DataFrame(data = vfile_rows)
```

```python
#convert date
vfiledf['reg_date'] = vfiledf['reg_date'].apply(lambda x: dt.datetime.strptime(x,'%m%d%Y'))
```

```python
vfiledf = vfiledf[['first','middle','last','suffix','birth_year','gender',
                   'reg_date','house_number_character','residence_street_number',
                   'house_suffix','predirection','street_name','street_type','suffix_direction',
                   'residence_extension','city','state','zip','mail_add_1','mail_add_2','mail_add_3',
                   'mail_add_4','mail_add_5','voter_id','county_code','jurisdiction','ward_precinct',
                   'school_code','state_house','state_senate','us_congress','county_commissioner',
                   'village_code','village_precinct','school_precinct','permanent_absentee_ind',
                   'status_type','uocava_status']]
```

## "H" file

```python
#open file
hfile = open('entire_state_h.lst')
hfile_lines = hfile.readlines()
```

```python
#test
sample_line = (hfile_lines[0])
print(sample_line)
```

```python
#separate each line into a dictionary
hfile_rows = []
for line in hfile_lines[:5]:
    row = {}
    row['voter_id'] = line[0:13]
    row['county_code'] = line[13:15]
    row['jurisdiction'] = line[15:20]
    row['school_code'] = line[20:25]
    row['election_code'] = line[25:38]
    row['absentee_voter_indicator'] = line[38:39]
    hfile_rows.append(row)
```

```python
#create dataframe
hfiledf = pd.DataFrame(data = hfile_rows)
hfiledf
```

# Combine Files

```python
df = pd.merge(hfiledf, vfiledf, on = ['voter_id'], how='outer')
df.head(100)
```
